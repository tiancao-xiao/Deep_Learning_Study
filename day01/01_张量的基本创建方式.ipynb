## 案例: 演示张量的基本创建方式.

张量(Tensor):

    PyTorch框架属于 最常用的深度学习框架, 无论是后续要学的ANN(人工神经网络), 还是CNN(卷积神经网络), RNN(循环神经网络)
    底层在处理数据时, 都是使用 张量 来处理的.

    张量 -> 存储同一类型元素的容器, 且元素值必须是 数值才可以.

张量的基本创建方式:

    torch.tensor 根据指定数据创建张量
    torch.Tensor 根据形状创建张量, 其也可用来创建指定数据的张量
    torch.IntTensor、torch.FloatTensor、torch.DoubleTensor 创建指定类型的张量

细节:

    Tensor方式较之于 tensor方式, 可以基于 形状来直接创建张量.

需要你掌握的方式:

    tensor(值, 类型), 例如:
        data = np.random.randint(0, 10, size=(2, 3))
        t3 = torch.tensor(data, dtype=torch.float)

#### 导包

```python
import numpy as np
import torch
```

#### 1. 定义函数，演示：torch.tensor 根据指定数据创建张量

```python
def dm01():
    # 场景1：标量 张量
    t1 = torch.tensor(10)   # 这种方式会创建一个包含值10的标量张量, 数据类型默认为int64
    print(f't1: {t1}, type: {type(t1)}')
    print('-' * 50)
    
    # 场景2：列表 -> 张量
    data = [[1, 2, 3], [4, 5, 6]]
    t2 = torch.tensor(data)
    print(f't2: {t2}, type: {type(t2)}')
    print('-' * 50)
    
    # 场景3：numpy nd数组 -> 张量
    data = np.random.randint(0, 10, size=(2, 3))
    t3 = torch.tensor(data, dtype=torch.float)  # 可以指定数据类型，这里把int转换为float
    print(f't3: {t3}, type: {type(t3)}')
    
    # 场景4：尝试直接创建指定维度（例如2行3列）的张量
    # t4 = torch.tensor(2, 3)  # 这种方式会报错, 因为torch.tensor期望的是数据而不是形状
    # print(f't4: {t4}, type: {type(t4)}')
```

```python
dm01()  # 掌握
```

#### 2. 定义函数，演示：torch.Tensor 根据形状创建张量, 其也可用来创建指定数据的张量

```python
def dm02():
    # 场景1：标量 张量
    t0 = torch.Tensor(10)   # 这种方式会创建尺寸为(10,)的一维张量, 数据类型默认为float32
    t1 = torch.Tensor([10]) # 这种方式会创建一个包含值10的标量张量, 数据类型默认为float32
    print(f't0: {t0}, type: {type(t0)}')
    print(f't1: {t1}, type: {type(t1)}')
    print('-' * 50)
    
    # 场景2：列表 -> 张量
    data = [[1, 2, 3], [4, 5, 6]]
    t2 = torch.Tensor(data)
    print(f't2: {t2}, type: {type(t2)}')
    print('-' * 50)
    
    # 场景3：numpy nd数组 -> 张量
    data = np.random.randint(0, 10, size=(2, 3))
    t3 = torch.Tensor(data)     # torch.Tensor会根据输入数据的类型自动推断张量的数据类型，不能直接指定dtype
    print(f't3: {t3}, type: {type(t3)}')
    print('-' * 50)
    
    # 场景4：尝试直接创建指定维度（例如2行3列）的张量
    t4 = torch.Tensor(2, 3)     # 这种方式会创建一个未初始化的张量，包含随机值  
    t5 = torch.Tensor((2, 3))  # 这种方式会创建一个包含值[2, 3]的张量
    print(f't4: {t4}, type: {type(t4)}')
    print(f't5: {t5}, type: {type(t5)}')
```

```python
dm02()
```

#### 3. 定义函数，演示：torch.IntTensor、torch.FloatTensor、
#### torch.DoubleTensor 创建指定类型的张量

```python
def dm03():
    # 场景1：标量 张量
    t1 = torch.IntTensor(10)
    print(f't1: {t1}, type: {type(t1)}')
    print('-' * 50)
    
    # 场景2：列表 -> 张量
    data = [[1, 2, 3], [4, 5, 6]]
    t2 = torch.IntTensor(data)
    print(f't2: {t2}, type: {type(t2)}')
    print('-' * 50)
    
    # 场景3：numpy nd数组 -> 张量
    data = np.random.randint(0, 10, size=(2, 3))
    t3 = torch.IntTensor(data)     
    print(f't3: {t3}, type: {type(t3)}')
    print('-' * 50)
    
    # 场景4：如果类型不匹配，会尝试自动转换类型
    data = np.random.randint(0, 10, size=(2, 3))
    t4 = torch.FloatTensor(data)        # 默认：float32
    print(f't4: {t4}, type: {type(t4)}')
```

```python
dm03()
```
